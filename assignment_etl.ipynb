```python
resear=soup.findAll(attrs = {'class':'container'})
```

```python
print(resear)
```

```python
rese=resear[0].text.replace('\n','')
```

```python
print(rese)
```

```python
summary=soup.find(attrs={'class':'title'})
```

```python
print(summary)
```

```python
id=int(soup.find(attrs={'class':'EL-header-and-subheader'}).text.replace('\n',''))  
```

```python
print(id)
```

```python
total_atrs=soup.find(attrs={'class':"col-xs-12 panel-menu"})
```

```python
total_atrs
```

```python
data=[['url','resear','rese','total_atrs','summary',]]
```

```python
data
```

```python
df=pd.DataFrame(data,columns=['url','resear','rese','total_atrs','summary'])
```

```python
df
```

```python

for i in range(1,10):
    url="https://fred.stlouisfed.org/categories/22"+str(i)
    page=requests.get(url)
    if(page.status_code==200):
        print("data feched sucessfully",i)
        soup=BeautifulSoup(page.content,'html.parser')
        #soup=BeautifulSoup(page.content,'html-parser')
        resear=soup.findAll(attrs= {'class':'container'})
        r=resear[0].text.replace('\n','')
        
        data.insert(id,[url,r,rese,total_atrs,summary])
    else:
        print("url not found")
```

```python
df=pd.DataFrame(data,columns=['url','resear','rese','total_atrs','summary'])
df.to_csv('inter.csv')
```
